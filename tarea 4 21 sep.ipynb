```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
import scipy as sp
```

```python
def f(theta): #definimos una funcion con la variable theta, para la integral
    return np.cos((m*(theta))-(x*(np.sin(theta))))   #valores de la integral
```

```python
#definimos variables
N = 1000
a = 0
b = np.pi
m = 5
x = np.arange(0,20,0.5) #tomamos una seccion de valores xd


def J(m,x): #definimos otra funcion para que una vez que se hace la integral la evalue
    
    h = (b-a)/N
    bessel = 0  #necesario definir la funcion inicial como 0, para que tenga valor
    xi = a #primer termino
    
    for i in np.arange (0, N, np.pi): #para hacer la suma de los cachitos N, tomamos de 0, en cachitos, hasta b
        integral=(h/3)*((f(xi))+(4*(f(xi+(2*i*h))))+(f(xi+2*(i+1)*h))+(f(b)))
        bessel += integral # equivalente i =i + 1 para una sucesion
        xi= xi + 2*h
    return bessel/np.pi
print(J(m,x))
```
