```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import math as mt
```

```python
#1.
#a)
def f(th):
    return np.cos(m*th-x*np.sin(th))


a=0
b=np.pi
N=1000
h=(b-a)/N
x=np.linspace(0,20,1000)
def J(m,x):
    sum0=0
    sum1=0
    for i in np.arange(0,N,2):
        sum0+=2*f(a+(2*i+1)*h)
        sum1+=4*f(a+(2*i)*h)
        integral = (h/3)*(f(a)+f(b)+sum0+sum1)

    bessel=(1/np.pi)*integral
    return bessel

m=0
j0=J(m,x)

m=1
j1=J(m,x)

m=2
j2=J(m,x)

plt.plot(x,j0)
plt.plot(x,j1)
plt.plot(x,j2)

plt.title("Funciones de Bessel",fontsize=14)
plt.xlabel("x",fontsize=12)
plt.ylabel("J(x)",fontsize=12)

plt.axhline(y = 0.0, xmin = 0.0, color = 'black')
plt.axvline(x = 0.0, ymin = 0.0, color = 'black')
plt.grid()
plt.show()
```

```python
#3
from scipy.constants import Boltzmann
V=0.001 #metro cubico
rho=6.022e28
TD=428
T=200
N=50
def gauss_xw_ab(N,a,b):
    x,w = sp.special.roots_legendre(N)
    return 0.5*(b-a)*x+0.5*(b+a),0.5*(b-a)*w

x, w = gauss_xw_ab(N,a,b)


def f(x):
    return ((x**4)*(np.e**x))/((np.e**x)-1)**2
a=0
b=TD/T


def CuadGauss(N,f,a,b):
    S = 0
    
    w,x = gauss_xw_ab(N,a,b)
    
    for i in range(N):
        S = S + w[i]*f(x[i])
        
    return S

CuadGauss(N,f,w,x)

Cv=(9*V*sp.constants.Boltzmann*rho*(T/TD)**3)*CuadGauss(N,f,w,x)

print(Cv)
```

```python
T=np.linspace(5,500)
cv=(9*V*sp.constants.Boltzmann*rho*(T/TD)**3)*CuadGauss(N,f,w,x)

plt.plot(T,cv)
plt.title("Cv vs T",fontsize=14)
plt.xlabel("Temperatura",fontsize=12)
plt.ylabel("Capacidad Calorifica",fontsize=12)

plt.axhline(y = 0.0, xmin = 0.0, color = 'black')
plt.axvline(x = 0.0, ymin = 0.0, color = 'black')
plt.grid()
plt.show()
```
