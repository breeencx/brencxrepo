A partir del programa para la $Eliminacion Gaussiana$ que hicimos en clase, escribe una programa que calcule la $descomposición$ $LU$ de una matriz. De hecho, el calculo es el mismo que para la $eliminación$ $gaussiana$, excepto que en cada paso del calculo es debes extraer los elementos apropiados de la matriz y ensamblarlos para formar la matriz diagonal inferior $L$ (i.e; $L$ = $L_0^1$ L_1^−1 L_2^−1 L^_3^-1 L_4^−1· · · ).

```python
import math as mt
import numpy as np
from math import *
import scipy as sp
import matplotlib.pyplot as plt
```

```python
1# 1a)
import numpy as np

N = 4

A = np.zeros([N,N])
U = np.zeros([N,N])
L = np.zeros([N,N])

print("digita los siguientes numeros y tras cada uno de ellos, presiona enter:")
print (" 2, 1, 4, 1, 3, 4, -1, -1, 1, -4, 1, 5, 2, -2, 1, 3")
for fila in range(0, N):
    for columna in range(0, N):
        A[fila, columna] = float(input("[" + str(fila+1)+ " , " +str(columna+1)+ "]" ))
        U[fila, columna] = A[fila, columna]
print(A)
for k in range(0, N):
    for fila in range(0, N):
        if (k == fila):
            L[k,fila] = 1
        if (k < fila):
            factor = (A[fila, k]/A[k, k])
            L[fila, k] = factor
            for columna in range (0, N):
                A[fila,columna]=A[fila,columna]-(factor*A[k,columna])
                U[fila,columna]=A[fila,columna]
print(U)
print(L)
```

```python
#1b)
v = np.array([ -4, 3, 9, 7 ],float)
N = len(v)

A = np.array([[ 2,  1,  4,  1 ],
           [ 3,  4, -1, -1 ],
           [ 1, -4,  1,  5 ],
           [ 2, -2,  1,  3 ]],float)

U = np.zeros([N, N])
L = np.zeros([N, N])

for fila in range(0, N):
    for columna in range(0, N):
        U[fila, columna] = A[fila, columna]
for k in range(0, N):
    for renglon in range(0, N):
        if (k == renglon):
            L[k, renglon] = 1
        if (k < renglon):
            factor = (A[renglon, k]/A[k, k])
            L[renglon, k] = factor
            for columna in range (0, N):
                A[renglon, columna] = A[renglon, columna]-(factor*A[k, columna])
                U[renglon, columna] = A[renglon, columna]

print("Dada la matriz A: ")
print(A)
print()
print("La matriz U será:")
print(U)
print()
print("La matriz L será:")
print(L)
```
