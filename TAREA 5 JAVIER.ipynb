```python
v = np.array([ -4, 3, 9, 7 ],float)

def LU(A, v):
    N = len(v)

    U = np.zeros([N, N])
    L = np.zeros([N, N])

    for renglon in range(0, N):
        for columna in range(0, N):
            U[renglon, columna] = A[renglon, columna]
    for k in range(0, N):
        for renglon in range(0, N):
            if (k == renglon):
                L[k, renglon] = 1
            if (k < renglon):
                factor = (A[renglon, k]/A[k, k])
                L[renglon, k] = factor
                for columna in range (0, N):
                    A[renglon, columna] = A[renglon, columna]-(factor*A[k, columna])
                    U[renglon, columna] = A[renglon, columna]
print(U)
print(L)

La = v[0]
Lb = v[1]-(L[1,0]*La)
Lc = v[2]-(L[2,0]*La)-(L[2,1]*Lb)
Ld = v[3]-(L[3,0]*La)-(L[3,1]*Lb)-(L[3,2]*Lc)

z = Ld/U[3,3]
y = (Lc-U[2,3]*z)/U[2,2]
x = (Lb-U[1,3]*z-U[1,2]*y)/U[1,1]
w = (La-U[0,3]*z-U[0,2]*y-U[0,1]*x)/U[0,0]

print("La solución del sistema de ecuaciones será")
print("w = ", w)
print("x = ", x)
print("y = ", y)
print("z = ", z)
```

<h3 align="center">$ y = {b\over{a+b^{2}}} $</h3>

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

def f(a, b, x, y):
    x0 = x
    y0 = y
    x = y0*(a+x0**2)
    y = b/(a+x0**2)
    return x, y

a = 1
b = 2

x = 1
y = 1

for n in range(70):
    x, y = f(a, b, x, y)
    print("n = ", n, "x = ", x, "y = ", y)
```
