```python
import numpy as np

# Cosntantes en eV
L = 6.199206681073*(10**(-16))
a = 10
M = 510999.6423167
h = 6.582119569 *(10**(-16))

# Definimos las constantes que multiplican a las integrales
B = (2*a)/(L**2)

# Definimos la función para calcular un elemento de la matriz H conociendo su m y n
def H(m,n):
    if m == n:
        return ((((h*np.pi*n)**2)/(M*(L**3)))*(L/2))+(B*((L**4)/4))
    elif m%2==0 and n%2==1:
        return (B*(-((2*L)/np.pi)**2)*((m*n)/((m**2-n**2)**2)))
    elif m%2==1 and n%2==0:
        return (B*(-((2*L)/np.pi)**2)*((m*n)/((m**2-n**2)**2)))
    else:
        return 0

print("Programa que calcula un elemento de la matriz H dada condiciones iniciales")
i = float(input("Ingrese la coordenada m de la matriz: "))
j = float(input("Ingrese la coordenada n de la matriz: "))
print(H(i,j))
```

Queándonos al final la siguiente ecuación

Para la primera, al igual que antes despejamos $y$

<h3 align="center">$ y = {x\over{a+x^{2}}} $</h3>

<h3 align="center">$ {x\over{a+x^{2}}} = {b\over{a+x^{2}}} $</h3>

Cancelando los denominadores

<h3 align="center">$ x = b $</h3>

Sustituyendo el valor de $x$ para la ecuación de la $y$

<h3 align="center">$ y = {b\over{a+b^{2}}} $</h3>

Para ver como se reorganizan de la segunda despejamos la $b$ del numerador 

<h3 align="center">$ y(a+b^{2}) = b $</h3>

y sustituimos $b$ para la primera, quedándonos 

<h3 align="center">$ x = y(a+b^{2}) $</h3>

recordando que $b = x$

<h3 align="center">$ x = y(a+x^{2}) $</h3>

Para la segunda simplemente tomamos la ecuación

<h3 align="center">$ y = {b\over{a+b^{2}}} $</h3>

y usamos que $x = b$ para la $b^{2}$, por lo tanto

<h3 align="center">$ y = {b\over{a+x^{2}}} $</h3>


```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

def f(a, b, x, y):
    x0 = x
    y0 = y
    x = y0*(a+x0**2)
    y = b/(a+x0**2)
    return x, y

a = 1
b = 2

x = 1
y = 1

for n in range(70):
    x, y = f(a, b, x, y)
    print("n = ", n, "x = ", x, "y = ", y)
```
