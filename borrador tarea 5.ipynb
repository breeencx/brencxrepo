```python
#PREELIMINAR 1C)

v = np.array([ -4, 3, 9, 7 ],float)

def LU(A, v):
    N = len(v)

    U = np.zeros([N, N])
    L = np.zeros([N, N])

    for renglon in range(0, N):
        for columna in range(0, N):
            U[renglon, columna] = A[renglon, columna]
    for k in range(0, N):
        for renglon in range(0, N):
            if (k == renglon):
                L[k, renglon] = 1
            if (k < renglon):
                factor = (A[renglon, k]/A[k, k])
                L[renglon, k] = factor
                for columna in range (0, N):
                    A[renglon, columna] = A[renglon, columna]-(factor*A[k, columna])
                    U[renglon, columna] = A[renglon, columna]
print(U)
print(L)
print(v)
La = v[0]
Lb = v[1]-(L[1,0]*La)
Lc = v[2]-(L[2,0]*La)-(L[2,1]*Lb)
Ld = v[3]-(L[3,0]*La)-(L[3,1]*Lb)-(L[3,2]*Lb)

print("La soluci칩n del sistema de ecuaciones ser치")
print("w = ", La)
print("x = ", Lb)
print("y = ", Lc)
print("z = ", Ld)
```
